```python
!pip install requests

import requests
import numpy as np
from ultralytics import YOLO
import cv2
from PIL import Image, ImageDraw, ImageFont

# Load the YOLOv8 model
model = YOLO("yolov8m.pt")

# Define the mapping from labels to URLs
label_to_url = {
    "cat": "https://www.amazon.in/s?k=dog&crid=N1Q6ZJSJI8ZH&sprefix=dog%2Caps%2C268&ref=nb_sb_noss_1",
    "dog": "https://www.amazon.in/s?k=dog&crid=N1Q6ZJSJI8ZH&sprefix=dog%2Caps%2C268&ref=nb_sb_noss_1",
    # Add more mappings as needed
}

# Load and process the image
image_path = "https://media.dev.to/cdn-cgi/image/width=800%2Cheight=%2Cfit=scale-down%2Cgravity=auto%2Cformat=auto/https%3A%2F%2Fdev-to-uploads.s3.amazonaws.com%2Fuploads%2Farticles%2Fwr8bm7gga15xp9gfz7yz.jpg"  # Replace with the path to your image

# Fetch the image from the URL
response = requests.get(image_path)
# Convert the image data to a NumPy array
image_array = np.asarray(bytearray(response.content), dtype="uint8")
# Decode the image using OpenCV
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for PIL
image_pil = Image.fromarray(image_rgb)

# Run the YOLOv8 model to detect objects
results = model(image)

# Process the results
result = results[0]

# Initialize drawing context
draw = ImageDraw.Draw(image_pil)

# Font for drawing text
try:
    font = ImageFont.truetype("arial.ttf", 15)
except IOError:
    font = ImageFont.load_default()

# Iterate over each detected box
for box in result.boxes:
    class_id = result.names[box.cls[0].item()]
    cords = box.xyxy[0].tolist()
    cords = [round(x) for x in cords]
    conf = round(box.conf[0].item(), 2)

    # Get the corresponding URL for the detected label
    url = label_to_url.get(class_id, "URL not found")

    # Draw the bounding box on the image
    x1, y1, x2, y2 = cords
    draw.rectangle([(x1, y1), (x2, y2)], outline="red", width=2)

    # Put the URL text instead of the label
    text_size = draw.textsize(url, font=font)
    text_background = [x1, y1 - text_size[1], x1 + text_size[0], y1]
    draw.rectangle(text_background, fill="red")
    draw.text((x1, y1 - text_size[1]), url, fill="white", font=font)

    # Print the details (optional)
    print("Object type:", class_id)
    print("Coordinates:", cords)
    print("Probability:", conf)
    print("Amazon URL:", url)
    print("---")

# Display the image using PIL
image_pil.show()
```

```python
!pip install requests ultralytics pillow

import requests
import numpy as np
from ultralytics import YOLO
import cv2
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display, HTML

# Load the YOLOv8 model
model = YOLO("yolov8m.pt")

# Define the mapping from labels to URLs
label_to_url = {
    "cat": "https://www.amazon.in/s?k=cat",
    "dog": "https://www.amazon.in/s?k=dog",
    # Add more mappings as needed
}

# Load and process the image
image_url = "https://media.dev.to/cdn-cgi/image/width=800%2Cheight=%2Cfit=scale-down%2Cgravity=auto%2Cformat=auto/https%3A%2F%2Fdev-to-uploads.s3.amazonaws.com%2Fuploads%2Farticles%2Fwr8bm7gga15xp9gfz7yz.jpg"  # Replace with the URL to your image

# Fetch the image from the URL
response = requests.get(image_url)
# Convert the image data to a NumPy array
image_array = np.asarray(bytearray(response.content), dtype="uint8")
# Decode the image using OpenCV
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for PIL
image_pil = Image.fromarray(image_rgb)

# Run the YOLOv8 model to detect objects
results = model(image)

# Process the results
result = results[0]

# Initialize drawing context
draw = ImageDraw.Draw(image_pil)

# Font for drawing text
try:
    font = ImageFont.truetype("arial.ttf", 15)
except IOError:
    font = ImageFont.load_default()

# Create a list to store HTML elements for hyperlinks
html_elements = []

# Iterate over each detected box
for box in result.boxes:
    class_id = result.names[int(box.cls[0])]
    cords = box.xyxy[0].tolist()
    cords = [round(x) for x in cords]
    conf = round(box.conf[0].item(), 2)

    # Get the corresponding URL for the detected label
    url = label_to_url.get(class_id, "URL not found")

    # Draw the bounding box on the image
    x1, y1, x2, y2 = cords
    draw.rectangle([(x1, y1), (x2, y2)], outline="red", width=2)

    # Put the URL text instead of the label
    draw.text((x1, y1 - 15), url, fill="red", font=font)

    # Add the hyperlink HTML element
    html_elements.append(f'<a href="{url}" target="_blank">{url}</a>')

# Display the image using PIL
display(image_pil)

# Display the hyperlinks below the image
html_content = "<br>".join(html_elements)
display(HTML(html_content))

```
